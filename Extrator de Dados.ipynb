```python
#define função para platagem

def plot_series(time, series, format="-", start=0, end=None, label=None):
    plt.plot(time[start:end], series[start:end], format, label=label)
    plt.xlabel("Time")
    plt.ylabel("Value")
    if label:
        plt.legend(fontsize=14)
    plt.grid(True)
```

```python
# Criar um dicionário para armazenar os dataframes de cada dia
dfs = {}

# Loop sobre as datas únicas da coluna "time"
for date in dates:
    # Filtrar as linhas correspondentes ao dia atual
    df_day = df_win.loc[df_win['time'].dt.date == date]
    # Adicionar o dataframe filtrado ao dicionário
    dfs[str(date)] = df_day
```

```python
for date in dates:
    # Filtrar as linhas correspondentes ao dia atual
    df_day = df_win.loc[df_win['time'].dt.date == date]
    # Nomear o dataframe com a data correspondente
    df_day_name = 'df_day_' + str(date)
    # Adicionar o dataframe filtrado ao dicionário com o nome correspondente
    dfs[df_day_name] = df_day
```

```python
# Criar um diretório para armazenar os arquivos CSV (se não existir)
if not os.path.exists('dataframes_csv2'):
    os.makedirs('dataframes_csv2')

# Percorrer o dicionário dfs e salvar cada dataframe em um arquivo CSV
for key, df in dfs.items():
    filename = key + '.csv'
    filepath = os.path.join('dataframes_csv2', filename)
    df.to_csv(filepath, index=False)
```
