```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd
import webbrowser
from datetime import datetime

# Dados do exemplo
data = pd.read_csv('C:/Users/adria/1Python/Meu_novo_projeto/WINJ23/WINJ23.csv')
#data["time"] = pd.to_datetime(data["time"], unit='s')
df = data

# Cria um subplot com dois eixos verticais (um para o gráfico de velas e outro para o volume)
fig = make_subplots(rows=2, cols=1, shared_xaxes=True, vertical_spacing=0.05)

# Plota o gráfico de velas no primeiro subplot
fig.add_trace(go.Candlestick(x=df['time'],
                open=df['open'],
                high=df['high'],
                low=df['low'],
                close=df['close'], name='WINJ23'), row=1, col=1)

# Plota o volume no segundo subplot
fig.add_trace(go.Bar(x=df['time'], y=df['real_volume'], name='Volume'), row=2, col=1)

# Define o layout do gráfico
fig.update_layout(title='Gráfico de candle com volume',
                  yaxis=dict(title='Preço'),
                  yaxis2=dict(title='Volume', side='right'),
                  xaxis_rangeslider_visible=False)

# Salva o gráfico em HTML
fig.write_html('grafico.html')

# Abre o arquivo HTML em uma nova janela do navegador
webbrowser.open_new_tab('grafico.html')
```
